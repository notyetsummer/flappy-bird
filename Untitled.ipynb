```python
import pygame
import pygame_menu
import random
import time

def set_difficulty(x,y):
    difficulty.append(y)

def MyTextValue(Player_name):
    name.append(str(Player_name))

def read_results():
    with open("Results.txt") as f:
        s = f.read()
        a = s.split(",")
    leaders = sorted(a, reverse=True)
    return leaders
    
def player_sprite():
    return [pygame.image.load("/Users/admin/flappy bird/frame_1.png"),pygame.image.load("/Users/admin/flappy bird/frame_2.png"),
                pygame.image.load('/Users/admin/flappy bird/frame_3.png'), pygame.image.load('/Users/admin/flappy bird/frame_4.png')
                ,pygame.image.load('/Users/admin/flappy bird/frame_5.png'),pygame.image.load('/Users/admin/flappy bird/frame_6.png')]



size = 1200, 800
width, height = size
GREEN = (150, 255, 150)
RED = (255, 0, 0)
Transparent = (0, 0, 0)
BLUE = (0, 0, 255)
name = ['John Doe']
difficulty = [1]


# Initialize Pygame
leaders = read_results()
pygame.init()




#/////////////////////////
def start_the_game():
    
    clock = pygame.time.Clock()
    t0 = time.time()
    
    value = 0
    end_flag = False
    width_wall = 100
    distece_walls = 500
    points = - width + 100
    space_time_0 = -1
    space_time_1 = 1
    running = True
    speed = [0, 1]  
    Y = -1
    real_points = 0

    # pygame
    Player_Sprite = player_sprite()
    sysfont = pygame.font.get_default_font()
    font = pygame.font.SysFont(None, 48)
    img = font.render(sysfont, True, RED)
    rect = img.get_rect()
    pygame.draw.rect(img, BLUE, rect, 1)
    font1 = pygame.font.SysFont(str(points), 72)
    img_points = font1.render(str(points), True, RED)
    screen = pygame.display.set_mode((width, height))
    pygame.display.set_caption('My Pygame Window')
    
    # pygame rects
    wall = pygame.Rect(width-100, size[1]-300, width_wall, size[1] - 200)
    ball = pygame.image.load("/Users/admin/flappy bird/main_frame.png")
    ball_hard = pygame.image.load("/Users/admin/Downloads/frame_1.png")
    end_screen = pygame.image.load("/Users/admin/Downloads/end_screen.jpg")
    rect = ball.get_rect()
    rect = rect.move([100,100])
    end_screen_rect = end_screen.get_rect()
    end_screen_rect = end_screen_rect.move([1,1])
    if difficulty[-1] == 1: ball = ball_hard     
    
    
    #Code for walls 
    
    wall_counter = int(width/(distece_walls + width_wall)) 

    delta0 = width - wall_counter*(distece_walls + width_wall)
    delta = distece_walls - delta0 - width_wall
    wall_counter = wall_counter + 1
    
    list_wall = [0]*wall_counter
    list_wall2 = [0]*wall_counter
    for i in range(wall_counter):
        R = random.randint(10,100)
        list_wall[i] = pygame.Rect(width-100 +i*distece_walls, height-300- R, width_wall, height -200)
        list_wall2[i] = pygame.Rect(width-100 +i*distece_walls, 0, width_wall, height - 700 - R )

    
    while running:
        clock.tick(120)

        #events
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    if rect.top > 0:
                        if difficulty[-1] == 1:
                            if space_time_0 == -1:
                                space_time_0 = time.time()
                                Y = rect.top
                                speed[1] = -3
                            else:
                                space_time_1 = time.time()
                                if (space_time_1 - space_time_0) > 1:
                                    Y = rect.top
                                    speed[1] = -3
                                    space_time_0 = space_time_1 
                        else:
                            Y = rect.top
                            speed[1] = -3
        #end events


        # moving
        if not end_flag:
            if Y!=-1 and Y - 150 > rect.top:
                speed[1] = 1
                Y = -1
            for i in range (wall_counter):
                if rect.colliderect(list_wall[i]):
                    with open("Results.txt", "a") as f: f.write(str(points)+ (" ") + (name[-1]) + (","))
                    end_flag = True
            for i in range (wall_counter):
                if rect.colliderect(list_wall2[i]):
                    with open("Results.txt", "a") as f: f.write(str(points) + (" ") + (name[-1]) + (","))
                    end_flag = True
                    return
            if rect.bottom < height: rect = rect.move(speed)
            for i in range(wall_counter):
                list_wall[i].move_ip(-1,0)
                list_wall2[i].move_ip(-1,0)
            points += 1
            if points == 0:
                real_points +=1
            if points > 0:
                real_points = points//(distece_walls + width_wall)
            img_points = font1.render(str(real_points), True, RED)
            # Teleporting Script
            for i in range(wall_counter):
                if list_wall[i].right < 10:
                    R = random.randint(-50,50) 
                    list_wall[i].move_ip([width+delta,R])
                    list_wall2[i].move_ip([width+delta,R])
        #endmoving


        #drawing
        if value >= len(Player_Sprite)*10:value = 0
        image = Player_Sprite[int(value/10)]
        screen.fill(GREEN)
        for i in range(wall_counter):
            pygame.draw.rect(screen, RED, list_wall[i])
            pygame.draw.rect(screen, BLUE, list_wall2[i])
        screen.blit(img_points, (20, 50))
        #screen.blit(ball, rect)
        screen.blit(image, rect)
        value += 1
        #if end_flag:screen.blit(end_screen, end_screen_rect)
        pygame.display.update() 
#////////////////////////////////////////////////////////////////////////////////////////////
    


screen = pygame.display.set_mode((width, height))
menu = pygame_menu.Menu('Welcome', width, height, theme=pygame_menu.themes.THEME_BLUE)
menu.add.text_input('Name :', default = "John Doe", onchange= MyTextValue)
menu.add.selector('Difficulty :', [('Hard', 1), ('Easy', 2)], onchange=set_difficulty)
menu.add.button('Play',start_the_game)
menu.add.button('Quit', pygame_menu.events.EXIT)
menu.add.button('------------------------------------------')
menu.add.button("Results:")
menu.add.button(leaders[0])
menu.add.button(leaders[1])
menu.add.button(leaders[2])
menu.add.button(leaders[3])
menu.add.button(leaders[4])
menu.add.button(leaders[5])
menu.mainloop(screen)


#//////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////


#//////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

# Quit Pygame
pygame.quit()
```
